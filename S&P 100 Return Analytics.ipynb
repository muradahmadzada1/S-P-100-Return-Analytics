<div class = "panel panel-primary">
    <div class = "panel-heading"> 
    <span style="color:white"><b>Task 10</b>
    </div>
    <div class = "panel-body"> 
    From the correlation matrix <code>corr_returns</code>, identify the pair of stocks that are <i>most</i> positively correlated (BAC and JPM) and the pair that are <i>most</i> uncorrelated (DUK and NFLX). 
    <br><br>Comment on whether it is reasonable for the identified pairs of stocks to be positively correlated or uncorrelated, and propose a reason as to why for each of the two pairs of stocks.
    <p style="text-align:right;">[15pts]</p>
    </div>
</div>

```python
def pairs(stock_returns):
    pairs_drop = set()
    cols = stock_returns.columns
    for i in range(0,stock_returns.shape [1] ):
        for j in range(0, i+1):
            pairs_drop.add((cols[i], cols[j]))
    return pairs_drop
def positive_corr(stock_returns, n=5):
    correlation = stock_returns.corr().abs().unstack()
    label_drop = pairs(stock_returns)
    correlation = correlation.drop(labels=label_drop).sort_values(ascending = False)
    return correlation[0:n]
print(positive_corr(stock_returns,1))
```

```python
def uncorrelated(stock_returns, n=5):
    correlation = stock_returns.corr().abs().unstack()
    label_drop = pairs(stock_returns)
    correlation = correlation.drop(labels=label_drop).sort_values(ascending = True)
    return correlation[0:n]
print(uncorrelated(stock_returns,1))
```

From the results it can be seen that BAC and JPM are positively correlated and the DUK and NFLX are weak positively correlated stocks. It means that BAC and JPM are moving the same direction while DUK and NFLX are moving same direction as well but as it is weak positively correlation it means they are not moving same direction as much as BAC and JPM. BAC(Bank Of America) and JPM(JPMorgan) are in the same industry so that's why it is so nomral that they have postive correlation and moving same direction. Something happens in the banking industry directly affect these two companies in a same manner. On the other hand, DUK(Duke Energy Corp) and NFLX(Netflix) are in the different industries : energy and broadcasting, respectively. So, that's why they have weak postive colleration, which means that if something happend in energy industry it will not affect NFLX but it will affect Duke Energy.
